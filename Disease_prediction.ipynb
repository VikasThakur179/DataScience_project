```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import mode
import warnings
warnings.filterwarnings("ignore")
```

```python
DATA_PATH = "Training.csv"
data = pd.read_csv(DATA_PATH).dropna(axis = 1)
data
```

```python
disease_counts = data["prognosis"].value_counts()
temp_df = pd.DataFrame({
    "Disease": disease_counts.index,
    "Counts": disease_counts.values
})
```

```python
plt.figure(figsize = (18,8))
sns.barplot(x = "Disease", y = "Counts", data = temp_df)
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data["prognosis"]=encoder.fit_transform(data["prognosis"])
data
```

```python
x=data.iloc[:,:-1]
x
```

```python
y=data.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print("shape of x_train",x_train.shape)
print("shape of y_train",y_train.shape)
print("shape of x_test",x_test.shape)
print("shape of y_test",y_test.shape)
```

```python
# Model building
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix , accuracy_score
```

```python
# Defining scoring metrics for k-fold cross valdiation
def cv_scoring(estimator, x,y):
    return accuracy_score(y, estimator.predict(x))
```

```python
# Initilizaing
models={"SVC":SVC(),
        "GaussianNB":GaussianNB(),
        "Random Forest":RandomForestClassifier(random_state=42)
       }
models       
```

```python
# Producing score for the cross valdiation model
for model_name in models:
    model=models[model_name]
    scores=cross_val_score(model,x,y, cv=10, n_jobs=-1,scoring=cv_scoring)

print("=="*30)
print(model_name)
print(f"scores: {scores}")
print(f"Mean score:{np.mean(scores)}")
```

```python
svm=SVC()
svm.fit(x_train,y_train)
pred=svm.predict(x_test)
print(f"accuracy score on train data by SVM classifier\: {accuracy_score(y_train,svm.predict(x_train))*100}")
print(f"Accuracy on test data by SVM Classifier\: {accuracy_score(y_test, pred)*100}")
```

```python
cm=confusion_matrix(y_test,pred)
plt.figure(figsize=(12,8))
sns.heatmap(cm,annot=True)
plt.title("CM for SVM classifier on Test data")
plt.show()
```

```python
nb=GaussianNB()
nb.fit(x_train,y_train)
pred=nb.predict(x_test)
print(f"accuracy on train data by NB classifier\: {accuracy_score(y_train, nb.predict(x_train))*100}")
print(f"accuracy on test data by NB classifier\: {accuracy_score(y_test, pred)*100}")
```

```python
cm=confusion_matrix(y_test,pred)
plt.figure(figsize=(12,8))
sns.heatmap(cm, annot=True)
plt.title("CM for NB on test data")
plt.show()
```

```python
rb=RandomForestClassifier(random_state=42)
rb.fit(x_train,y_train)
pred=rb.predict(x_test)
print(f"Accuracy on train data RFC\: {accuracy_score(y_train, rb.predict(x_train))*100}")
print(f"Accuracy on test data RFC\:{accuracy_score(y_test,pred)*100}")
```

```python
cm = confusion_matrix(y_test,pred)
plt.figure(figsize=(12,8))
sns.heatmap(cm ,annot=True)
plt.title("CM for RFC on Test data")
plt.show()
```

```python
# Training the model whole data

final_svm_model=SVC()
final_nb_model=GaussianNB()
final_rb_model=RandomForestClassifier(random_state=42)

final_svm_model.fit(x,y)
final_nb_model.fit(x,y)
final_rb_model.fit(x,y)
```

```python
# Reading the test data
test_data=pd.read_csv("Testing.csv").dropna(axis=1)
X=test_data.iloc[:,:-1]
Y=encoder.transform(test_data.iloc[:,-1])
```

```python
# Making prediction by mode of prediction made by all the classifiers
svm_pred=final_svm_model.predict(X)
nb_pred=final_nb_model.predict(X)
rb_pred=final_rb_model.predict(X)
```

```python
from scipy import stats
final_pred=[stats.mode([i,j,k])[0] for i,j,k in zip(svm_pred,nb_pred,rb_pred)]
print(f"accuracy od train data set by combined model: {accuracy_score(Y,final_pred)*100}")

cf = confusion_matrix(Y,final_pred)
plt.figure(figsize=(10,8))
sns.heatmap(cf,annot=True)
plt.title("Confusion matrix for combined model on test dataset")
plt.show()
```

```python
symptoms=x.columns.values
symptom_index ={}
for index, value in enumerate(symptoms):
    sym = " ".join([i.capitalize() for i in value.split("_")])
    symptom_index[sym] = index

data_dict={
    "symptom_index":symptom_index,
    "pred_classes":encoder.classes_
}
```

```python
# Defining the Function
# Input: string containing symptoms separated by commas
# Output: Generated predictions by models

def predictions(symptoms):
    symptoms=symptoms.split(",")


    # Creating input for the model
    input_data =[0]* len(data_dict["symptom_index"])
    for symptom in symptoms:
        index =data_dict["symptom_index"][symptom]
        input_data[index]=1

    # reshaping the input data and converting it into suitable 
# format for model predictions.

    input_data = np.array(input_data).reshape(1,-1)

# generating individual output data
    rb_pred=data_dict["pred_classes"][final_rb_model.predict(input_data)[0]]
    svm_pred = data_dict["pred_classes"][final_svm_model.predict(input_data)[0]]
    nb_pred = data_dict["pred_classes"][final_nb_model.predict(input_data)[0]]
# making final prediction by taking mode all predictions 
# Use statistics.mode instead of scipy.stats.mode
    import statistics
    final_prediction = statistics.mode([rb_pred, nb_pred, svm_pred])
    predictions = {
        "rb_model_prediction": rb_pred,
        "naive_bayes_prediction": nb_pred,
        "svm_model_prediction": svm_pred,
        "final_prediction":final_prediction
    }
    return predictions
 #testing the function
print(predictions("Itching,Skin Rash,Nodal Skin Eruptions"))
```
